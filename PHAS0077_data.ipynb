Six datasets for rural hospital closures have been collected. Now, we are going to merge them all into one big dataset by unique CCNs. 

```python
df1 = pd.read_csv('/Users/tonydong/Desktop/PHAS0077/Data/hosp_info.csv')
df2 = pd.read_csv('/Users/tonydong/Desktop/PHAS0077/Data/hosp_eff.csv')
df3 = pd.read_csv('/Users/tonydong/Desktop/PHAS0077/Data/hosp_pc.csv')
df4 = pd.read_csv('/Users/tonydong/Desktop/PHAS0077/Data/hosp_mort_read.csv')
df5 = pd.read_csv('/Users/tonydong/Desktop/PHAS0077/Data/hosp_pat_exp.csv')
df6 = pd.read_csv('/Users/tonydong/Desktop/PHAS0077/Data/hosp_external_fac.csv')
df7 = pd.merge(df1,df2,on='PRVDR_NUM')
df8 = pd.merge(df7,df3,on='PRVDR_NUM')
df9 = pd.merge(df8,df4,on='PRVDR_NUM')
df10 = pd.merge(df9,df5,on='PRVDR_NUM')
df11 = pd.merge(df10,df6,on='PRVDR_NUM')
# Delete duplicates
df11.drop_duplicates(subset=['PRVDR_NUM'],keep='last',inplace=True)

print(df11.shape)
df11.head(100)
```

```python
# Find #N/A
for column in df11:
    count_nan = df11[column].isna().sum()
    nan_pct   = count_nan / len(df11[column]) * 100
    print(f"[column] has {round(nan_pct, 2)}% nan")
    if nan_pct > 40:
        df11 = df11.drop(column, axis = 1)
```

```python
df11.drop(['CLOSURE_YEAR','HOSP_TYPE','ZIP_CD','FIPS_CD','CBSA_CD'],axis=1,inplace=True)
print(f"the shape of df11 is: {df11.shape}")
list(df11)
```
